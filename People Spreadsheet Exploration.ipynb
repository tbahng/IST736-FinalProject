```python
df.groupby('Party').sum()
```

```python
df_party = df.groupby('Party').sum()
df_party.reset_index(inplace=True)
subset = ['Party', 'Governor', 'Senate', 'House']
df_party = df_party[subset]
#sns.set()
df_party.set_index('Party').T.plot(kind='bar', stacked=True)
```
